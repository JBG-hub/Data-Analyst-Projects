```python
import pandas as pd
df = pd.read_csv(r"C:\Users\jgauc\Jupyter Notebooks\Pandas\Portfolio Project\dirty_cafe_sales.csv")
pd.set_option('display.min.rows', 100)
df
```

```python
# Start Data Cleaning. First drop duplicate rows, if any.
df = df.drop_duplicates()
```

```python
# Explore columns looking for data to clean. 
# Using the .unique() function to visualize the values in each column
for col in df:
    print(col)
    print(df[col].unique())
    print('\n')
```

```python
# For our three quantitative columns, convert anything that's Error, Unknown or nan to 0: 
# Before: array(['2', '4', '5', '3', '1', 'ERROR', 'UNKNOWN', nan] --> after:  array([2., 4., 5., 3., 1., 0.])
# The reason we have to cast these to 0 is because some functions and operators will not work when columns contain str and numeric
df['Quantity'] = df['Quantity'].apply(pd.to_numeric, errors='coerce').fillna(0)
df['Price Per Unit'] = df['Price Per Unit'].apply(pd.to_numeric, errors='coerce').fillna(0)
df['Total Spent'] = df['Total Spent'].apply(pd.to_numeric, errors='coerce').fillna(0)
```

```python
# Cast all nulls (nan) in the remaining columns to empty strings
df = df.fillna('')
```

```python
# First look at Item and Price Per Unit pairs.
print(df.groupby(['Item', 'Price Per Unit']).size().reset_index(name='count'))
```

```python
# Create a dictionary that pairs the Item and the Unit Price.
dictionary_item_prices = {'Cake':  3.0, 'Coffee':  2.0, 'Cookie':  1.0, 'Juice':  3.0, 'Salad':  5.0, 
                          'Sandwich':  4.0, 'Smoothie':  4.0, 'Tea':  1.5}

# Given the Item name, we can use this dictionary to determine missing unit prices.
for index, row in df.iterrows():
        if row['Price Per Unit'] == 0 and dictionary_item_prices.get(row['Item']):
            df.at[index, 'Price Per Unit'] =  dictionary_item_prices.get(row['Item'])
```

```python
# Checkpoint. Still have more clean-up to do. Next we will need to resolve as many of the blanks, Errors and Unknowns as possible.
print(df.groupby(['Item', 'Price Per Unit']).size().reset_index(name='count'))
```

```python
# We can fill in missing Quantity values if given both 1) unit price and 2) total price.
# Fill in Quantity values with the below code.
for index, row in df.iterrows():
        if row['Quantity'] == 0 and row['Price Per Unit'] > 0 and  row['Total Spent'] > 0:
            df.at[index, 'Quantity'] =  row['Total Spent'] / row['Price Per Unit']

# Next fill in Unit Price with similar logic.
for index, row in df.iterrows():
        if row['Price Per Unit'] == 0 and row['Quantity'] > 0 and  row['Total Spent'] > 0:
            df.at[index, 'Price Per Unit'] =  row['Total Spent'] / row['Quantity']

# Next fill in Total Spent
for index, row in df.iterrows():
        if row['Total Spent'] == 0 and row['Quantity'] > 0 and  row['Price Per Unit'] > 0:
            df.at[index, 'Total Spent'] =  row['Quantity'] * row['Price Per Unit']
```

```python
# Checkpoint - Looking much cleaner for Qty, Unit Price and Total Spent.
df.head(100)
```

```python
print(df.groupby(['Item', 'Price Per Unit']).size().reset_index(name='count'))
```

```python
# Keep cleaning up these pairs.
# Some items are the only item sold at a particular price point (i.e. Cookie is the only item sold at $1). 
# Below is a dictionary of those items.
# Use the dictionary to fill in some missing Item values.
dictionary_unique_prices = {2.0: 'Coffee', 1.0: 'Cookie',  5.0: 'Salad', 1.5: 'Tea'}

for index, row in df.iterrows():
        if not dictionary_item_prices.get(row['Item']) and dictionary_unique_prices.get(row['Price Per Unit']):
            df.at[index, 'Item'] =  dictionary_unique_prices.get(row['Price Per Unit'])
            
```

```python
print(df.groupby(['Item', 'Price Per Unit']).size().reset_index(name='count'))
# Better. Next we can consolidate blanks, ERROR and UNKNOWN into a single value
```

```python
# For every column, replace each instance of blank or ERROR with UNKNOWN to consolidate values.
dictionary_unknowns = {'': 'UNKNOWN', 'ERROR':'UNKNOWN'}

for colname, colitems in df.items():
    for index, row in colitems.items():
        var = df[colname][index]
        if dictionary_unknowns.get(var):
            df.at[index, colname] = dictionary_unknowns.get(var)
```

```python
# Checkpoint. We have filled in as much data using logic as we can.
print(df.groupby(['Item', 'Price Per Unit']).size().reset_index(name='count'))
```

```python
df.head(100)
```

```python
# Next step of Data Cleaning will be to drop rows where qty, unit price or total spent = 0.
# These rows will not be able to give us meaningful business insights.
df = df[(df['Quantity']>0)]
df = df[(df['Price Per Unit']>0)]
df = df[(df['Total Spent']>0)]
```

```python
# Rows dropped successfully; no rows remain under those conditions
df[(df['Quantity']==0) | (df['Price Per Unit']==0) | (df['Total Spent']==0)]
```
